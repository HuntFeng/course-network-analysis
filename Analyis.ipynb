# Load data and create NetworkX graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
from pyvis.network import Network
from make_nx_graph import make_nx_graph
```

```python
g = make_nx_graph(["math"])
```

```python
# draw network
nt = Network(500, 500, notebook=True)
nt.from_nx(g)
nt.show("graph.html")
```

# Start analysis

```python
degree = pd.Series(dict(g.degree))
plt.figure(figsize=(15,6))
degree.plot.bar(x='Courses', y='Degree')
plt.title("Total degree of each vertex")
```

```python
nx.node_connectivity(g)
```

```python
betweenness = nx.betweenness_centrality(g)
betweenness = pd.Series(betweenness)
plt.figure(figsize=(15,6))
betweenness.plot.bar(x="Courses", y="Betweenness")
plt.title("Betweenness of each vertex")
```

```python
nx.average_clustering(g)
```

```python
nx.incidence_matrix(g)
```

```python
nx.adjacency_spectrum(g)
```
